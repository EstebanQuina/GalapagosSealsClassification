```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import InceptionV3 # Use InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input # Import its specific preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils import class_weight # To calculate class weights

# 1. Mount Google Drive (essential if in a new session)
print("Mounting Google Drive...")
from google.colab import drive
drive.mount('/content/drive')
print("Google Drive mounted successfully!")

# 2. Define the ROOT directory for your ORIGINAL IMBALANCED dataset on Google Drive
ORIGINAL_IMBALANCED_DATA_ROOT = os.path.join('/content/drive/My Drive', 'vgg16_seals_classification_dataset')

# Verify that the dataset exists
if not os.path.exists(ORIGINAL_IMBALANCED_DATA_ROOT):
    print(f"Error: Original imbalanced dataset not found at {ORIGINAL_IMBALANCED_DATA_ROOT}.")
    print("Please ensure the initial 'Phase 1: Data Preparation for VGG16' was run successfully.")
    exit()

# 3. Define dataset paths for training, validation, and test splits
TRAIN_DIR_IMBALANCED = os.path.join(ORIGINAL_IMBALANCED_DATA_ROOT, 'train')
VAL_DIR_IMBALANCED = os.path.join(ORIGINAL_IMBALANCED_DATA_ROOT, 'val')
TEST_DIR_IMBALANCED = os.path.join(ORIGINAL_IMBALANCED_DATA_ROOT, 'test')

# 4. Define your class names
CLASS_NAMES = [
    "Arctocephalus_galapagoensis", # Minority class
    "Zalophus_wollebaeki"        # Majority class
]
NUM_CLASSES = len(CLASS_NAMES)
TARGET_IMG_SIZE = (299, 299) # InceptionV3's common input size
BATCH_SIZE = 32 # Revert to original batch size as dataset is larger again

print(f"\nTraining InceptionV3 model on ORIGINAL IMBALANCED data (Oversampling Case) for {NUM_CLASSES} classes: {CLASS_NAMES}")
print(f"Using dataset from: {ORIGINAL_IMBALANCED_DATA_ROOT}")
print(f"Cropped images resized to: {TARGET_IMG_SIZE}")


# 5. Data Generators with Augmentation (for training) and Preprocessing
# Aggressive augmentation for training to help minority class
# IMPORTANT: Use InceptionV3's specific preprocess_input
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input, # InceptionV3 preprocessing
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input) # InceptionV3 preprocessing
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input) # InceptionV3 preprocessing

train_generator = train_datagen.flow_from_directory(
    TRAIN_DIR_IMBALANCED,
    target_size=TARGET_IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASS_NAMES,
    shuffle=True
)

val_generator = val_datagen.flow_from_directory(
    VAL_DIR_IMBALANCED,
    target_size=TARGET_IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASS_NAMES,
    shuffle=False
)

test_generator = test_datagen.flow_from_directory(
    TEST_DIR_IMBALANCED,
    target_size=TARGET_IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    classes=CLASS_NAMES,
    shuffle=False
)

print("\nData generators created.")
print(f"Found {train_generator.samples} training images belonging to {train_generator.num_classes} classes.")
print(f"Found {val_generator.samples} validation images belonging to {val_generator.num_classes} classes.")
print(f"Found {test_generator.samples} test images belonging to {test_generator.num_classes} classes.")


# --- Calculate Class Weights ---
print("\nCalculating class weights...")
train_labels = train_generator.classes
class_weights_dict = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(train_labels),
    y=train_labels
)
class_weights_dict = {i: class_weights_dict[i] for i in range(len(class_weights_dict))}
print(f"Calculated Class Weights: {class_weights_dict}")


# 6. Build the InceptionV3 Model for Transfer Learning
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=(TARGET_IMG_SIZE[0], TARGET_IMG_SIZE[1], 3))

for layer in base_model.layers:
    layer.trainable = False

x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# 7. Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()

# 8. Define Callbacks for training
model_checkpoint_path_inceptionv3_oversampled = os.path.join( # NEW PATH for InceptionV3 Oversampled
    '/content/drive/My Drive',
    'inceptionv3_training_results',
    'inceptionv3_classification_oversampled_case', # Specific subfolder
    'best_model.h5'
)
os.makedirs(os.path.dirname(model_checkpoint_path_inceptionv3_oversampled), exist_ok=True)

checkpoint = ModelCheckpoint(
    model_checkpoint_path_inceptionv3_oversampled,
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)

early_stopping = EarlyStopping(
    monitor='val_accuracy',
    patience=10,
    mode='max',
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',
    factor=0.2,
    patience=5,
    min_lr=0.000001,
    mode='max',
    verbose=1
)

callbacks_list = [checkpoint, early_stopping, reduce_lr]

# 9. Train the Model
EPOCHS_INCEPTIONV3_OVERSAMPLED = 50

print("\nStarting InceptionV3 model training (Oversampled Case)...")
history_inceptionv3_oversampled = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // BATCH_SIZE,
    epochs=EPOCHS_INCEPTIONV3_OVERSAMPLED,
    validation_data=val_generator,
    validation_steps=val_generator.samples // BATCH_SIZE,
    callbacks=callbacks_list,
    class_weight=class_weights_dict # <-- CRUCIAL for oversampling strategy
)

print("\nInceptionV3 Training (Oversampled Case) complete!")
print(f"Best model saved to: {model_checkpoint_path_inceptionv3_oversampled}")

# Save training history
history_inceptionv3_oversampled_dict = history_inceptionv3_oversampled.history
history_filepath_inceptionv3_oversampled = os.path.join(os.path.dirname(model_checkpoint_path_inceptionv3_oversampled), 'training_history.json')
with open(history_filepath_inceptionv3_oversampled, 'w') as f:
    json.dump(history_inceptionv3_oversampled_dict, f)
print(f"Training history saved to: {history_filepath_inceptionv3_oversampled}")


# --- 10. Evaluate on test set and get more metrics ---
print("\nEvaluating model on the test set for detailed metrics...")

Y_pred_proba_inceptionv3_oversampled = model.predict(test_generator, steps=test_generator.samples // BATCH_SIZE + (test_generator.samples % BATCH_SIZE != 0))
Y_pred_classes_inceptionv3_oversampled = np.argmax(Y_pred_proba_inceptionv3_oversampled, axis=1)

Y_true_inceptionv3_oversampled = test_generator.classes[test_generator.index_array]

print("\n--- InceptionV3 Classification Report (Oversampled Case) ---")
report_inceptionv3_oversampled = classification_report(Y_true_inceptionv3_oversampled, Y_pred_classes_inceptionv3_oversampled, target_names=CLASS_NAMES)
print(report_inceptionv3_oversampled)

# Save Classification Report to Drive
report_filepath_inceptionv3_oversampled = os.path.join(os.path.dirname(model_checkpoint_path_inceptionv3_oversampled), 'classification_report.txt')
with open(report_filepath_inceptionv3_oversampled, 'w') as f:
    f.write(report_inceptionv3_oversampled)
print(f"Classification report saved to: {report_filepath_inceptionv3_oversampled}")

print("\n--- InceptionV3 Confusion Matrix (Oversampled Case) ---")
cm_inceptionv3_oversampled = confusion_matrix(Y_true_inceptionv3_oversampled, Y_pred_classes_inceptionv3_oversampled)

plt.figure(figsize=(8, 6))
sns.heatmap(cm_inceptionv3_oversampled, annot=True, fmt='d', cmap='Blues', xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('InceptionV3 Confusion Matrix (Oversampled Case)')
plt.show()

# Save Confusion Matrix plot to Drive
cm_filepath_inceptionv3_oversampled = os.path.join(os.path.dirname(model_checkpoint_path_inceptionv3_oversampled), 'confusion_matrix.png')
plt.savefig(cm_filepath_inceptionv3_oversampled)
print(f"Confusion matrix plot saved to: {cm_filepath_inceptionv3_oversampled}")
plt.close()


overall_test_loss_inceptionv3_oversampled, overall_test_accuracy_inceptionv3_oversampled = model.evaluate(test_generator, steps=test_generator.samples // BATCH_SIZE)
print(f"\nOverall Test Loss: {overall_test_loss_inceptionv3_oversampled:.4f}")
print(f"Overall Test Accuracy: {overall_test_accuracy_inceptionv3_oversampled:.4f}")
```
